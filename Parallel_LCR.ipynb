```python
import numpy as np
import matplotlib.pyplot as plt
plt.rcParams['font.family'] = 'serif'
plt.rcParams['font.serif'] = ['Times New Roman'] + plt.rcParams['font.serif']
```

```python
F =np.array([100,200,300,400,500,600,700,750,800,850,900,950,1000,1100,1200,1500,2000,2500,3000,3500,4000,4500,5000,5500,6000])
f =np.log10(F)
V =np.array([0.362,0.438,0.501,0.539,0.562,0.577,0.586,0.588,0.589,0.591,0.596,0.593,0.591,0.588,0.587,0.583,0.56,0.536,0.504,0.471,0.44,0.405,0.375,0.344,0.316])
X_l = 2*np.pi*F*0.0623
X_c = 1/(2*np.pi*F*0.47*10**(-6))
Z= 1/(np.sqrt((1/(993**2))+((1/X_c) - (1/X_l))**2))
I =V/Z
```

```python
plt.plot(f,V)
plt.scatter(f,V)
plt.xlabel('Log of Frequency (Hz)')
plt.ylabel('Output Voltage (V)')
plt.scatter(2.9542425094393248,0.596, color ='cyan', marker = 's') #experimental resonant frequency
plt.scatter(2.96848295,0.593, color ='magenta', marker = 's') #theoretical resonant frequency

plt.axvline(x=2.240, ymin=0, ymax=0.955, color='green', linestyle= '--') #bandwidth
plt.axvline(x=3.625, ymin=0, ymax=0.955, color='green', linestyle= '--') #bandwidth
plt.axhline(y=0.596, xmin=0, xmax=4, color='red', linestyle= '--')
plt.axvline(x=2.9542425094393248, ymin=0, ymax=0.955, color='cyan', linestyle= '--')
plt.axhline(y=0.596/np.sqrt(2), xmin=0, xmax=4, color='red', linestyle= 'dashdot')
plt.show()
print(np.amax(V))

plt.scatter(f,I)
plt.axvline(x=2.9542425094393248, ymin=0, ymax=1, color='cyan', linestyle= '--')
plt.axhline(y=0.0006011515968873986, xmin=0, xmax=4, color='red', linestyle= '--')
plt.xlabel('Log of Frequency (Hz)')
plt.ylabel('Output Current (A)')
plt.show()
print(np.amin(I))
#'Theoretical Resonant Frequency = 930.0947208 Hz'
```
