```python
import numpy as np
import matplotlib.pyplot as plt
plt.rcParams['font.family'] = 'serif'
plt.rcParams['font.serif'] = ['Times New Roman'] + plt.rcParams['font.serif']
```

PART A

```python
vp1 = np.array([7.759110651,8.584280466,13.37048099,27.27257231,34.53053578,53.71578976,64.78467195,57.42835812])
vp2 = np.array([7.501544849,8.333342014,13.03046196,27.68510956,35.10142117,54.31941172,65.28054313,57.9605144])
```

```python
plt.scatter(vp1,vp2)
plt.plot(vp1,vp2)
plt.xlabel('Velocity through Photogate 1 (cm/s)')
plt.ylabel('Velocity through Photogate 2 (cm/s)')
plt.show()
m,c = np.polyfit(vp1, vp2, 1)
print('m=', m, 'c=', c)
```

PART B

```python
Mass2 = np.array([173.2142857,115.8095238, 52.29787234,24.18269231, 20.44354839,15.88271605, 6.967821782]) #24.18269231,20.44354839,
Ai2 = np.array([0.195589841,0.118574841, 0.055690577,0.02655055, 0.023096816, 0.018332351, 0.006989382]) #0.02655055,0.023096816,
plt.scatter(Mass2,Ai2)
#plt.plot(Mass,Ai)
plt.xlabel('(M+m)/m')
plt.ylabel('1/Acceleration (s2/cm)')
m2,c2 = np.polyfit(Mass2, Ai2, 1)
print('m=', m2, 'c=', c2, 'g=', 1/m2)
Ai2n = [m2*x +c2 for x in Mass2]
plt.plot(Mass2, Ai2n)
plt.show()
```

```python
M = np.log(Mass2)
A = np.log(Ai2)
plt.scatter(M,A)
m3,c3 = np.polyfit(M, A, 1)
print('m=', m3, 'c=', c3, 'g=', 1/(np.exp(c3)))
Ain = [m3*i +c3 for i in M]
plt.plot(M, Ain)
plt.xlabel('(M+m)/m [logarithmic scale]')
plt.ylabel('1/Acceleration (s2/cm) [logarithmic scale]')
plt.show()
```
